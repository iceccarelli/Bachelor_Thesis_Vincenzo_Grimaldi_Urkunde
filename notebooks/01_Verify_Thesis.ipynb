# Verification of the Bachelor's Thesis Content

## Development of a Cross-domain Knowledge Base for Cyber Intelligence in Smart Grids

The Bachelor's thesis submitted at RWTH Aachen University in January 2021.



```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from pypdf import PdfReader

# Define paths to the thesis PDF and extracted metadata
BASE_PATH = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
THESIS_PATH = os.path.join(BASE_PATH, 'BA_Ceccarelli_Complete_1.pdf')
KEY_DETAILS_CSV = os.path.join(BASE_PATH, 'data', 'key_details.csv')

def extract_text_from_pdf(pdf_path):
    """Extract full text from the thesis PDF for verification."""
    full_text = ""
    try:
        reader = PdfReader(pdf_path)
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                full_text += page_text + "\n"
    except Exception as e:
        print(f"Error reading PDF: {e}")
    return full_text.lower()  # Normalize to lowercase for reliable matching

# Extract complete text once
thesis_text = extract_text_from_pdf(THESIS_PATH)

print("PDF Text Extraction Complete")
print("=" * 50)
print(f"Total characters extracted: {len(thesis_text)}")
print("First 800 characters for reference:")
print(thesis_text[:800])

# Load the structured metadata
key_details_df = pd.read_csv(KEY_DETAILS_CSV)
print("\nMetadata Loaded from key_details.csv")
print(key_details_df[['section', 'detail_key', 'detail_value']].head(10))
```

### Structured Verification Process

The verification proceeds section by section. For each recorded detail the notebook searches the full PDF text to confirm presence of the exact value. This covers:

- General author and university information
- Thesis titles, supervisors, and dates
- Chapter structure (titles and starting pages)
- Key facts from the introduction (investment, IoT projection, smart grid goals)
- Core technologies (Neo4j, Cypher, SCADA systems, cloud layers)
- Contributions in cross-domain modeling and threat detection
- Primary threats addressed (Advanced Persistent Threats and specific attack vectors)
- Abbreviations and core concepts used throughout the work

```python
def verify_detail(section, detail_key, expected_value):
    """Check if the expected value appears in the thesis text."""
    if pd.isna(expected_value) or str(expected_value).strip() == "":
        return True
    search_term = str(expected_value).strip().lower()
    if search_term in thesis_text:
        print(f"[VERIFIED] {section} → {detail_key}: '{expected_value}' confirmed in thesis.")
        return True
    else:
        print(f"[ISSUE] {section} → {detail_key}: '{expected_value}' not found in PDF text.")
        return False

# Run verification for every entry
verification_results = []
for _, row in key_details_df.iterrows():
    section = row['section']
    key = row['detail_key']
    value = row['detail_value']
    result = verify_detail(section, key, value)
    verification_results.append(result)

# Summary statistics
total_entries = len(key_details_df)
passed = sum(verification_results)
print("\n" + "=" * 60)
print(f"VERIFICATION SUMMARY: {passed}/{total_entries} details confirmed")
print("=" * 60)
```

### Additional Content Verification

Beyond metadata, the notebook checks for key technical elements that appear throughout the thesis, including references to Neo4j, Cypher queries, SCADA systems, specific threats, and example figures.

```python
# Specific technical terms verification
key_terms_to_check = {
    "Neo4j": "graph database and caching layers",
    "Cypher": "query language for relationship traversal",
    "SCADA": "Supervisory Control and Data Acquisition",
    "Advanced Persistent Threats": "APT detection focus",
    "DDoS": "flood attack examples",
    "Man-in-the-Middle": "MitM threat vector",
    "Figure 2.1.1": "European Electric Grid scheme",
    "Similarity Algorithms": "link analysis for anomalies"
}

print("\nTechnical Content Verification")
print("=" * 40)
for term, description in key_terms_to_check.items():
    count = thesis_text.count(term.lower())
    if count > 0:
        print(f"[CONFIRMED] '{term}' appears {count} times — {description}")
    else:
        print(f"[MISSING] '{term}' not detected in PDF")

# Quick visualization of verification coverage by section
section_counts = key_details_df.groupby('section').size()
verified_by_section = key_details_df.groupby('section').apply(
    lambda g: sum(verify_detail(g['section'].iloc[0], g['detail_key'].iloc[i], g['detail_value'].iloc[i]) for i in range(len(g)))
)

plt.figure(figsize=(10, 5))
plt.bar(section_counts.index, section_counts.values, alpha=0.7, label='Total entries')
plt.bar(verified_by_section.index, verified_by_section.values, alpha=0.9, label='Verified')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Number of Details')
plt.title('Verification Coverage by Thesis Section')
plt.legend()
plt.tight_layout()
plt.show()
```
