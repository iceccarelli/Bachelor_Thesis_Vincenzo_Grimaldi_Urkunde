# 02_Expertise_Analysis.ipynb: Analysis of My Bachelor's Thesis

This notebook walks through the core ideas, methods, and results from my Bachelor's thesis at RWTH Aachen University (January 2021). 

The work focused on building a **cross-domain knowledge base** using a graph database (Neo4j) for cyber intelligence in Smart Grids. The goal was to collect IoT network traffic passively, store it as relationships in a graph, and then use link analysis and similarity algorithms to detect anomalies and Advanced Persistent Threats (APTs) — especially relevant for the distribution grid side where SCADA, RTUs, and IEDs are exposed.

Everything here is taken directly from the thesis PDF and the extracted `key_details.csv`. No speculation — just the actual research and its practical value for real infrastructure like energy networks and electrified rail systems.

### Core Methodology (Chapter 4)

The practical approach was straightforward and production-oriented:

1. **Passive IoT traffic collection** from network gateways (headers, ports, DNS, MAC addresses, data volume, protocols like SSDP/mDNS).
2. **Store everything in Neo4j** as a graph (nodes = devices/events, edges = relationships).
3. **Run link analysis** to explore connections between entities (exactly as described in 4.1.1).
4. **Apply similarity algorithms** on the graph to spot anomalies (Section 4.3).
5. **Interpret the blackbox dataflow** (Section 4.2) to flag potential APTs or compromised devices in real time.

This matches the main objective in the metadata: “Create a feasible graph-based knowledge base using Neo4j to detect Advanced Persistent Threats (APTs) in Smart Grids through cross-domain data modeling, link analysis and cyber intelligence.”

The advantage over traditional databases is clear (Chapter 3): Neo4j handles complex relationships and traversals with caching layers far better, especially for the high-volume, unstructured data coming from RTUs, MTUs, IEDs, and PLCs in Smart Grids.

### Why This Matters for Real Infrastructure

The knowledge base I built directly supports:

- **Smart Grid security**: Modeling relationships between devices, traffic patterns, and events in transmission and (especially) distribution grids.
- **Renewables integration**: Monitoring volatile DERs and IoT sensors while keeping the system resilient against attacks.
- **Critical infrastructure protection**: The same graph approach can be applied to railway OT/IT systems (SCADA-style control, remote terminals, sensors) to detect anomalies before they escalate.

Everything was kept practical: passive collection (no extra load on the grid), Neo4j caching for speed, Cypher queries for analysis, and similarity scoring to flag compromised IoT devices. The thesis ends with a clear path for future work (Chapter 5.2) — scaling the model and adding more real-time triggers.
