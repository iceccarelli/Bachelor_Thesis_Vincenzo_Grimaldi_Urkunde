# Analysis of the Bachelor's Thesis

## Development of a Cross-domain Knowledge Base for Cyber Intelligence in Smart Grids

The thesis submitted at RWTH Aachen University in January 2021 examines the development of a cross-domain knowledge base that supports cyber intelligence within smart grid environments. The work addresses the challenges of handling high volumes of unstructured data generated by network protocols, IoT devices, and supervisory control systems in modern energy networks.

Electricity forms a central part of contemporary infrastructure, and the research highlights the need for scalable, self-healing systems that integrate renewables and distributed energy resources while maintaining resilience against cyber threats. The approach centers on modeling relationships between devices, traffic patterns, and events using a graph database, enabling efficient link analysis and similarity-based detection of anomalies in distribution grids.

All content below draws directly from the thesis structure, metadata, chapters, technologies, threats, and contributions as recorded in the extracted details.

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Load the complete metadata extracted from the thesis
BASE_PATH = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
KEY_DETAILS_CSV = os.path.join(BASE_PATH, 'data', 'key_details.csv')
key_details_df = pd.read_csv(KEY_DETAILS_CSV)

# Display the full metadata in a clean table format
print("Thesis Metadata Overview")
print("=" * 50)
print(key_details_df[['section', 'detail_key', 'detail_value']].to_string(index=False))

# Group and summarize by section for easier reference
grouped = key_details_df.groupby('section').apply(lambda x: x[['detail_key', 'detail_value']].to_dict('records'))
for section, items in grouped.items():
    print(f"\n{section.upper()} SECTION SUMMARY:")
    for item in items:
        print(f"  - {item['detail_key']}: {item['detail_value']}")
```

### Thesis Structure

The document is organized across five chapters, beginning with the motivation for improved scalability and security in energy systems and progressing through theoretical foundations, modeling techniques, practical implementation, and conclusions.

Chapter 1 introduces the urgency of modernizing electric infrastructure, referencing the estimated investment requirements and the projected growth of IoT-connected devices. It outlines the objective of constructing a graph-based knowledge base to support real-time threat detection.

Chapter 2 provides the theoretical background on network protocols, smart grid architectures, information and communication technologies, and the distinction between transmission and distribution grids. It covers SCADA systems, remote terminal units, master terminal units, intelligent electronic devices, and programmable logic controllers.

Chapter 3 explores graph models, the NoSQL approach, the advantages of Neo4j for handling complex relationships, Cypher query language, caching layers, and the traverser API.

Chapter 4 details the operational approach, including passive data collection, blackbox dataflow interpretation, link analysis for exploring entity relationships, and similarity algorithms for anomaly detection.

Chapter 5 summarizes the findings and indicates directions for continued development in scalable monitoring and automated response mechanisms.

```python
# Extract and display chapter details
chapters_df = key_details_df[key_details_df['section'] == 'Chapters'].copy()
print("Detailed Chapter Breakdown")
print(chapters_df[['detail_key', 'detail_value']].to_string(index=False))

# Extract technologies
tech_df = key_details_df[key_details_df['section'] == 'Technologies'].copy()
print("\nCore Technologies and Systems")
print(tech_df[['detail_key', 'detail_value']].to_string(index=False))

# Simple visualization of chapter progression
chapter_pages = [10, 14, 23, 38, 47]
chapter_names = ['Ch1 Intro', 'Ch2 Background', 'Ch3 Models', 'Ch4 Approach', 'Ch5 Conclusion']

plt.figure(figsize=(10, 4))
plt.barh(chapter_names, chapter_pages, color='#1f77b4')
plt.xlabel('Starting Page')
plt.title('Thesis Chapter Progression')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

### Core Methodology

The research implements a structured process for handling network traffic in smart grid environments. Data is collected passively from IoT devices and gateways, capturing headers, ports, DNS information, MAC addresses, data volumes, and protocols such as SSDP and mDNS.

This information is stored in a graph database using nodes for devices and events, with edges representing relationships. Link analysis examines connections between entities, while similarity algorithms evaluate patterns to identify deviations from normal behavior.

The blackbox dataflow interpretation focuses on real-time processing to flag potential advanced persistent threats. Neo4j provides efficient traversals and caching layers that support high-volume operations without the overhead of multiple table joins required in traditional relational structures.

This methodology aligns with the stated objective of creating a feasible graph-based knowledge base for detecting threats through cross-domain modeling and cyber intelligence techniques.

```python
# Extract threats addressed in the research
threats_df = key_details_df[key_details_df['section'] == 'Threats'].copy()
print("Threats Addressed in the Analysis")
print(threats_df[['detail_key', 'detail_value']].to_string(index=False))

# Extract contributions
contributions_df = key_details_df[key_details_df['section'] == 'Contributions'].copy()
print("\nKey Contributions from the Research")
print(contributions_df[['detail_key', 'detail_value']].to_string(index=False))

# Basic summary count
print(f"\nTotal entries in metadata: {len(key_details_df)}")
print(f"Sections covered: {key_details_df['section'].nunique()}")
```

### Practical Value for Infrastructure

The knowledge base supports secure operation of energy networks by modeling relationships across devices, traffic flows, and events in both transmission and distribution grids. It enables monitoring of volatile distributed energy resources and IoT sensors while maintaining system resilience.

The same principles apply to electrified rail systems, where operational technology and information technology components manage signaling, power supply, and station operations. Graph-based modeling facilitates anomaly detection in SCADA-style environments, supporting risk assessment and compliance in critical infrastructure.

Passive collection avoids additional load on operational systems, while caching and Cypher queries ensure responsive analysis. The approach contributes to scalable integration of renewables and robust protection against common attack vectors in high-stakes energy and transport settings.
